This notebook prepares the raw datasets for model training by applying a minimal preprocessing pipeline.

The following steps are performed:

- Load the raw datasets.
- Select the required columns.
- Apply text preprocessing.
- Save the processed datasets.

Import Libraries

```python
import pandas as pd
import numpy as np
```

Load Train, Test, Validation Data

```python
train_df = pd.read_csv('/content/drive/MyDrive/PROYECT_5/train.csv')
test_df = pd.read_csv('/content/drive/MyDrive/PROYECT_5/test.csv')
validation_df = pd.read_csv('/content/drive/MyDrive/PROYECT_5/validation.csv')
```

```python
print(train_df.shape)
print(test_df.shape)
print(validation_df.shape)
```

```python
train_df.head()
```

Select just Required Features

```python
train_processed = train_df[['argument', 'stance_WA']].copy()
test_processed = test_df[['argument', 'stance_WA']].copy()
validation_processed = validation_df[['argument', 'stance_WA']].copy()
```

```python
train_processed.head()
```

```python
test_processed.head()
```

```python
validation_processed.head()
```

Convert Text to Lowercase

```python
train_processed['argument'] = train_processed['argument'].str.lower()
test_processed['argument'] = test_processed['argument'].str.lower()
validation_processed['argument'] = validation_processed['argument'].str.lower()
```

```python
train_processed.head()
```

Normalize Whitespace

```python
train_processed['argument'] = train_processed['argument'].str.strip()
train_processed['argument'] = train_processed['argument'].str.replace(r'\s+', ' ', regex=True)
```

```python
test_processed['argument'] = test_processed['argument'].str.strip()
test_processed['argument'] = test_processed['argument'].str.replace(r'\s+', ' ', regex=True)
```

```python
validation_processed['argument'] = validation_processed['argument'].str.strip()
validation_processed['argument'] = validation_processed['argument'].str.replace(r'\s+', ' ', regex=True)
```

```python
train_processed.sample(n=5, random_state=42)
```

Save Processed Data

```python
train_processed.to_csv('/content/drive/MyDrive/PROYECT_5/train_processed.csv', index=False)
test_processed.to_csv('/content/drive/MyDrive/PROYECT_5/test_processed.csv', index=False)
validation_processed.to_csv('/content/drive/MyDrive/PROYECT_5/validation_processed.csv', index=False)
```

Summary

- Loaded the raw training, validation and test datasets.
- Selected only the columns required for model training.
- Converted all text to lowercase.
- Normalized whitespace by removing extra spaces.
- Saved the processed datasets for the modelling stage.
